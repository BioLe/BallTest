Uma overview de todos os dados representados num só gráfico. Abaixo é apresentada uma análise dos mesmos.

```python
df.plot(x='Time.time', y='PlayerPosX', figsize=(10,5), grid=True)
```

Eixo x representa o tempo.
Eixo y representa a posição do jogador em X.

É possível observar que o jogador tem uma área de teste com 20 metros (10m a -10m) de comprimento em X para percorrer.

Aos segundos 5, 6.2 e 12. Existe uma pequena oscilação que podemos identificar como uma colisão contra os limites da área de teste.

Aos segundos 5.1 e 14, a posição em X retorna a 0 de forma instanea. Sabemos então que o utilizador carregou no botão para colocar a bola na posição inicial.
